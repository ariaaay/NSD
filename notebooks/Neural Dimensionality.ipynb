```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
visual_roi_names = {
    -1: "non_cortical",
    0: "cortical",
    1: "V1v",
    2: "V1d",
    3: "V2v",
    4: "V2d",
    5: "V3v",
    6: "v3d",
    7: "h4v",
}

place_roi_names = {
    -1: "non_cortical",
    0: "cortical",
    1: "OPA",
    2: "PPA",
    3: "RSC",
}

face_roi_names = {
    -1: "non_cortical",
    0: "cortical",
    1: "OFA",
    2: "FFA-1",
    3: "FFA-2",
    4: "aTL-faces",
}

roi_name_dict = {
    "floc-faces": face_roi_names,
    "floc-places": place_roi_names,
    "prf-visualrois": visual_roi_names,
}
```

```python
output_dir = "/user_data/yuanw3/project_outputs/NSD/output"
```

```python
# rois
face_dict = roi_name_dict["floc-faces"]
plt.figure(figsize=(10,10))
for k, v in face_dict.items():
    if k > 0:
        plt.subplot(2, 2, k)
        for d in np.linspace(5, 100, 6):
            train_loss = np.load("%s/autoencode/train_loss_dim%d_%s_%s.npy" % (output_dir, d, "floc-faces", v))
            plt.plot(train_loss, label="Dim=%d" % d)
            plt.ylabel("MSE loss")
            plt.xlabel("Epochs")
            plt.legend()
            plt.title(v)
```

```python
# rois
face_dict = roi_name_dict["floc-faces"]
plt.figure(figsize=(5,5))
for k, v in face_dict.items():
    if k > 0:
        dims = np.linspace(5, 100, 6)
        test_losses = []
        for d in dims:
            test_losses.append(np.load("%s/autoencode/test_loss_dim%d_%s_%s.npy" % (output_dir, d, "floc-faces", v)))
        plt.plot(dims, test_losses, label=v)
plt.legend()
plt.ylabel("MSE loss")
plt.xlabel("dim")
```
